```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
df=pd.DataFrame(pd.read_csv("Diwali Sales Data.csv",encoding="unicode_escape"))
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df["Amount"]=df["Amount"].astype("int")
```

```python
df["Amount"].dtypes
```

```python
df.describe()
```

```python
df
```

```python
sns.set(style='whitegrid', context='talk', font_scale=0.7, 
        rc={'figure.figsize': (17, 5), 'axes.spines.top': False,'axes.labelsize':14,'axes.titlesize':18})
```

```python
ax=sns.countplot(x="Gender",data=df,hue="Gender",palette="Set2")
for bars in ax.containers:
    ax.bar_label(bars)
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Count by Gender")
```

```python
amt_by_gen=df.groupby("Gender", as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Gender",y="Amount",data=amt_by_gen,hue="Gender",palette="Pastel1")
plt.xlabel("Gender")
plt.ylabel("Amount")
plt.title("Amount by Gender")
```

*From the above two graph we can see that most of the buyers are females and even the purchasing power of females are greater than men.*

From the above two graph we can see that most of the buyers are females and the most amount of money is also spend by females.

```python
ax=sns.countplot(x="Age Group",data=df,hue="Gender",palette="Pastel2")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.title("Count By Age Group")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
grp_by_age=df.groupby("Age Group",as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Age Group",y="Amount",data= grp_by_age,hue="Age Group",palette ="Paired")
plt.xlabel("Age Group")
plt.ylabel("Amount")
plt.title("Amount by Age Group")
```

From the above two graph we can see that most of the buyers are of age group between 26-35 yrs female

From the above two graph we can see that most of the buyers are from 26-35 age group in which most of them are females

```python
sales_by_state=df.groupby("State",as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
ax1=sns.barplot(x="State",y="Orders",data=sales_by_state,hue="State",palette="muted")
plt.xlabel("State")
plt.ylabel("Orders")
plt.title("State Vs Orders")
for bars in ax1.containers:
   ax1.bar_label(bars)
```

```python
Amount_by_state=df.groupby("State",as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.barplot(x="State",y="Amount",data=Amount_by_state,hue="Amount",palette="Accent")
plt.xlabel("State")
plt.ylabel("Amount")
plt.title("Amount by State")
```

From the above two graph we can see that most of the buyers and the most amount of money is spend by the UP,Maharashtra and Karnataka. 

```python
ax2=sns.countplot(x="Marital_Status",data=df,hue="Marital_Status",palette="Accent",legend=False)
sns.set(style="whitegrid",context="talk",rc={"figure.figsize":(8,5),"axes.spines.top":False})
for bars in ax2.containers:
  ax2.bar_label(bars)
```

```python
amt_by_marital=df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Marital_Status",y="Amount",data=amt_by_marital,hue="Gender",palette="Set2")
```

```python
sns.set(rc={"figure.figsize":(30,10)},style="whitegrid",context="talk")
ax3=sns.countplot(x="Occupation",data=df,hue="Occupation",palette="Pastel2")
for bars in ax3.containers:
  ax3.bar_label(bars)
```

```python
amt_by_occ=df.groupby("Occupation",as_index=False)["Amount"].sum().sort_values(by="Occupation",ascending=False)
sns.barplot(x="Occupation",y="Amount",data=amt_by_occ,hue="Occupation",palette="Pastel1")
```

```python
amt_by_cate=df.groupby("Product_Category",as_index=False)["Amount"].sum().sort_values(by="Product_Category",ascending=False)
sns.barplot(x="Product_Category",y="Amount",data=amt_by_cate,hue="Product_Category",palette="Paired")
```
